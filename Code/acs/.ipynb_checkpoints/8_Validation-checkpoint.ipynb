## aggregating blocks

```python
prediction = pd.read_csv('../../Data/predicted/nn_1.csv',index_col=0).set_index('LOGRECNO')
```

```python
prediction.head(5)
```

```python
## need population for weight
blockdata = pd.read_csv('../../Data/refined_data/blocks50.csv',index_col='LOGRECNO')
```

```python
blockdata.head(5)
```

```python
weights = blockdata['our37862'] # Household owner occupied, total
```

```python
prediction['weight'] = weights
```

```python
prediction.head(2)
```

## LOGRECNO-GEOID bridge

```python
bridge = pd.read_excel('../../Data/boundaries/ny_logrecno_geoid.xls', index_col='LOGRECNO')[['GEOID']]
```

```python
bridge.head(2)
```

```python
p = prediction.merge(bridge,  left_index=True, right_index=True)
```

```python
p.head()
```

```python
p.GEOID = p.GEOID.str.replace('15000US','1500000US')
```

```python
p2 = p.merge(aggKey, how='left', on = 'GEOID' )
```

```python
p2.head(10)
```

```python
#### Now I need to generate weighted average per Pums
```

```python
def wavg(group):
    d = group['internet']
    w = group['weight']
    return (d * w).sum() / w.sum()

pPumas = p2.groupby('pumaID').apply(wavg)
```

```python
pPumas.head(5)
```

## Now I need to read real values

```python
real = pd.read_csv('../../Data/refined_data/predict.csv', index_col=0)[['internetP']]
real['st'] = real.index
real = real[real.st.str.startswith('ny')]
real['LOGRECNO'] = real.st.str.split('_').str.get(1).apply(lambda x: x.zfill(7))
real.drop('st', axis=1, inplace=1)
real.head(5)
```

```python
real.shape
```

## Bridge 2

```python
bridge_real = pd.read_excel('../../Data/boundaries/1_year_Mini_Geo.xls', sheetname='NY')
bridge_real.LOGRECNO = bridge_real.LOGRECNO.astype(str).apply(lambda x: x.zfill(7))
bridge_real.head(2)
```

```python
fig, ax = plt.subplots()
real2.internetP.hist(bins=20, ax=ax, alpha=.5, label='real')
(pPumas).hist(bins=20, ax=ax, alpha=.5, label='predicted')
plt.legend();
```

```python
predict = pd.DataFrame(pPumas).reset_index()
predict.columns = ('GEOID','predict')
predict.predict = predict.predict
```

```python
print predict.shape
print real2.shape
```

```python
real2.GEOID = real2.GEOID.str.replace('79500US','7950000US')
```

```python
check = predict.merge(real2, on='GEOID', how='left')
```

```python
check.head(10)
```

```python
def rsquared(df, predict, real):
    '''
    generate R2 metric, comparing all predicted values to all real
    '''
    return 1 - ((df[real] - df[predict])**2).sum()/((df[real] - df[real].mean())**2).sum()
```

```python
rsquared(check, 'predict', 'internetP')
```

## Error percentage distribution

```python
check['error_percent'] = ((check.internetP - check.predict)/check.internetP).abs()
```

```python
check.error_percent.hist(bins=50);
```

```python
1.0*check[check['error_percent']<.05].shape[0]/check.shape[0]
```

```python
1.0*check[check['error_percent']<.1].shape[0]/check.shape[0]
```

95% of PUMAS had error below 10%, and 68 - below 5%

## Mapping the error

```python
error_map = pumas[['AFFGEOID10', 'geometry']].merge(check[['GEOID','error_percent']], 
                                        how='left', left_on='AFFGEOID10', right_on='GEOID')
```

```python
fig, ax = plt.subplots(figsize=(18,18))
error_map.plot(column = 'error_percent', cmap='Reds', ax=ax)
# plt.axis('off')
```

```python
fig, ax = plt.subplots(figsize=(18,18))
error_map.plot(column = 'error_percent', cmap='Reds', ax=ax)
ax.set_xlim(-74.5, -73.5)
ax.set_ylim(40.5, 41.5)
plt.axis('off')
```
