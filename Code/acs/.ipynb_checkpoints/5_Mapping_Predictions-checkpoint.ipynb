```python
def save_Geojson(x, path, land=True):
    x.to_crs(epsg=4326)

    if land:
        x = x[x.AWATER < x.ALAND]  # drop water groups
    with open(path, 'w') as writer:
        writer.write(x.to_crs(epsg=4326).to_json())
        print 'json is at: %s' % path
        
        
```
