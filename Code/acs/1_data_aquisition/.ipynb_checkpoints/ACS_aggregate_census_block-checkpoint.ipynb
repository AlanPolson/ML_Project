```python
import os
import pandas as pd

import zipfile #i am not giving it a name cause i intend to use it only once
```

```python
path = '../../../Data/ACS/5y2013_NYS/NewYork/'
```

```python
def getFiles(folder, format, full=True, filtr=None):
    for f in os.listdir(folder):
        if f.endswith(format):
            if filtr==None or filtr not in f:
                if full:
                    yield os.path.join(folder, f)
                else:
                    yield f
```

```python
zips = list(getFiles(path, '.zip'))
zips[:10]
```

```python
def estimates(f):
    return 'e' + f.replace('.zip','.txt')
    
```

```python
cols = [
'FILEID',
'FILETYPE',
'STUSAB',
'CHARITER',
'SEQUENCE',
'LOGRECNO',
]
```

```python
def getTable(fname):
    zf = zipfile.ZipFile(fname)
    df = pd.read_csv(zf.open(estimates(fname.split('/')[-1])), 
                     dtype={5:str, 4:str},
                     header=None, low_memory=False)
    seq = df.iloc[1,4]

    df.columns = cols + ['%s_%d' % (seq , i+1) for i in xrange(df.shape[1] - 6)]
    return df.iloc[:,5:].set_index('LOGRECNO')
```

```python
t1 = getTable(zips[0])
```

```python
t1.head()
```

```python
def aggTables(folder):
    
    tables =(getTable(f) for f in getFiles(folder, '.zip'))
    return pd.concat(tables,1)
```

```python
mdf = aggTables(path)

mdf.head(3)
```
